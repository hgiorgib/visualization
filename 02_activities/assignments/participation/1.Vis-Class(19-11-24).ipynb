#### Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

#### Make figures: Scatter, Bar, and Line Plots

```python
# Make a figure
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100,50)

# Make a scatterplot
fig, ax = plt.subplots(figsize=(5, 3))      
ax.scatter(x,y)
```

```python
# Keeping the sane figure definition, change the axis object - Bar Plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
# Line Plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
# Histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
# Add title and labels to lineplot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years since start')
```

```python
# Add modified labels and titles to lineplot
# Make dictionaries with properties for lables
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

# Add fontdict arguments to previous lineplot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

#### Modify Data Points - Markers (Scatter Plot)

```python
# Modify data points (maker shape and colours) - Scatter Plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,y,
                marker='*', # NEW ADDITIONS         It can used the marker shapes: >, o, +
                color = "indigo") # NEW ADDITIONS
fig.show()
```

#### Modify Data Points - Line Attributes (Line Plot)

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,
            marker='*',
            color = 'indigo',
            linestyle = '--', # NEW ADDITIONS
            linewidth = 2) # NEW ADDITIONS
fig.show()
```

#### Modify Data Points - Colour (Line Plot)

*Colour names:*
https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallerycolor-named-colors-py

*Also, hex codes can be used for colours*

```python
fig, ax = plt.subplots(figsize=(5, 3))
# color being the new addition
ax.plot(x,y,marker='*', color = '#7425b9', linestyle = '--', linewidth = 2)
fig.show()
```

#### Modify Data Points - Marker and Marker Edge Colours (Line Plot)

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
                        markersize = 12, # NEW ADDITIONS
                        color = '#7425b9', # (indigo)
                        linestyle = '--',
                        linewidth = 2,
                        markeredgecolor = '#fa9359', # NEW ADDITIONS (orange)
                        markerfacecolor = '#000000' ) # NEW ADDITIONS (black)
```

#### Gridlines

Veritical Gridlines: ax.grid(axis = 'x')

Horizontal Gridlines: ax.grid(axis = 'y')

Both Gridlines: ax.grid()

```python
# Example 1
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
                        markersize = 12, # NEW ADDITIONS
                        color = '#7425b9', # (indigo)
                        linestyle = '--',
                        linewidth = 2,
                        markeredgecolor = '#fa9359', # NEW ADDITIONS (orange)
                        markerfacecolor = '#000000' ) # NEW ADDITIONS (black)

# Gridline settings
ax.grid(axis = 'y', color = "dodgerblue", linewidth = 2, linestyle = '-.')
```

```python
# Example 2
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*',
                        markersize = 12, # NEW ADDITIONS
                        color = '#7425b9', # (indigo)
                        linestyle = '--',
                        linewidth = 2,
                        markeredgecolor = '#fa9359', # NEW ADDITIONS (orange)
                        markerfacecolor = '#000000' ) # NEW ADDITIONS (black)

# Gridline settings
ax.grid()
```
