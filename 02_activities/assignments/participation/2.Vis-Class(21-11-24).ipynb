## Customize Plots

#### Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

#### Legends

```python
# Define sample data - with 2y sample variables (y1 and y y2)
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)

# Make a line plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

#### Modify Legend (for both sample variables y1 and y2)

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
            frameon = True,     #add frame around the legend
            fontsize = 12,      #change font size
            ncol = 2,           #specify number of columns
            shadow = True )
fig.show()
```

#### Moving Legend Outside Ploting Area

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
bbox_to_anchor =(1, 1))
fig.show()
```

## Text and Annotations

#### Scatter plot using the same sample data

```python
# Make scatter plot with labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# Graph title
ax.text(10, 95, "This value is important!")
fig.show()
```

```python
# Modify text annotations
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# Graph title
ax.text(10, 95, "This value is important!",
        ha='center',    ##alignment of text
        color = 'red',  ##modify font colour
        size = 20)      ##modify font size
fig.show()
```

### Positioning text annotations - Using Transforms

```python
# Make a line plot
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData )
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes )
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure )
```

#### Annotating with arrows

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
            arrowprops=dict(facecolor='black', shrink=0.05))
fig.show()
```

##### Different arrow style

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
            arrowprops=dict(facecolor='hotpink', arrowstyle = "wedge"))
fig.show()
```

### Axis Labels

##### Removing tick marks and labels

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator())       # Locator to remove ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter())   # Formatter to remove only labels
```

##### Limiting tick marks

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))   # Specifies 3 locators in X-axe (Only 3 values)
```

##### Tick mark intervals

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))   # Steps taken for the X-axe markers
```

##### Rotating axis labels

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, # 45 degrees rotation
            ha='right',
            fontsize = 12,
            font = "serif") 
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.yticks(rotation=45, # 45 degrees rotation
            ha='right',
            fontsize = 12,
            font = "serif") 
```

##### Modifying axis title fonts

```python
font1 = {'family':'serif','color':'indigo'}
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1)
```
